```python
# load data
data = ascii.read('comacluster_mushotzky.csv',format='csv')
print(data.keys())
```

### a) Sample galaxy cluster X-ray spectrum

The read in data are stored as a dictionary with colum headings 'ekev' (photon energy in units of keV) and photon number intensity 'nekev' (in units of photons / cm^2 / s / keV)

First, convert these quantities to flux density in cgs units. Then, make a log-log plot of the SED, $\nu L_\nu$ vs. $\nu$. (Hint: it might be useful to check your work by looking up a typical luminosity of a galaxy cluster)

The distance to this Galaxy cluster is $\simeq 100$ Mpc $= 3.1\times10^{26}$ cm, and its radius is $R \approx 1$ Mpc

```python
nu_cgs = data['ekev']*kev_to_Hz
photon_flux_cgs = data['nekev']/kev_to_Hz # photons/cm^2/s/Hz
distance_cgs = 3.1e26;
radius_cgs = 3.1e24;
energy_per_photon = h_cgs*nu_cgs
flux_density_cgs = photon_flux_cgs*energy_per_photon # erg/cm^2/s/Hz
Lnu_cgs = 4*np.pi*distance_cgs**2*flux_density_cgs # erg/s/Hz
plt.loglog(nu_cgs, nu_cgs*Lnu_cgs, marker='o', ls='');
plt.xlabel(r'$\nu$ [Hz]')
plt.ylabel(r'$\nu L_\nu$ [erg/s]');

# Typical galaxy cluster luminosity \approx 1e44 erg/s (https://ned.ipac.caltech.edu/level5/Sept01/Bahcall2/paper.pdf section 2.1)
```

### b) Estimate the color temperature $T_c$ from the X-ray spectrum

```python
def wiens_law(peak_nu):
    # peak_nu in Hz
    temp = peak_nu/(5.9e10) # K 
    return temp

peak_nu = 2e18;
color_temp = wiens_law(peak_nu)
print("Color temperature: {:.2e} K".format(color_temp))
```

### c) Compare to a one zone spherical bremsstrahlung model solution

now we can plot a model against the data using an emissivity appropriate for Bremsstrahlung from hydrogen with number density $n$ and temperature $T$, both in cgs units

$j_\nu \simeq 10^{-38} \, T^{-1/2} \, n^2 e^{-h\nu/kT}$ erg cm$^{-2}$ s$^{-1}$ Hz$^{-1}$ ster$^{-1}$

use your solution to Problem 1 in the optically thin limit, $\alpha_\nu R \ll 1$, along with this form of $j_\nu$ to estimate $n$ and $T$ for the intracluster medium

```python
def nuLnu(freq, temp, density):
    # all in cgs
    jnu = 1e-38/np.sqrt(temp)*density**2*np.exp(-h_cgs*freq/(kb_cgs*temp))
    Inu = 2.0*jnu*radius_cgs # here, taking b=0, max value of Inu
    Fnu = np.pi*radius_cgs**2/distance_cgs**2*Inu # estimate assuming constant emitter
    Lnu = 4*np.pi*distance_cgs**2*Fnu
    nuLnu = freq*Lnu
    return nuLnu

# Fit best density and temperature to data
from scipy.optimize import curve_fit
fitted_vals, _ = curve_fit(nuLnu, nu_cgs, nu_cgs*Lnu_cgs, p0=[1e7, 1e-3])
temp_fit, density_fit = fitted_vals
dense_freqs = np.logspace(np.log10(nu_cgs[0]), np.log10(nu_cgs[-1]), 1000)
fit_vals = nuLnu(dense_freqs, temp_fit, density_fit)

fit_string = 'Fit: T={:.2e} K, n={:.2e} particles/cm3'.format(temp_fit, density_fit)

# Typical number density of intracluster medium: 1e-3 particles/cm^3. See https://en.wikipedia.org/wiki/Intracluster_medium
```

```python
plt.loglog(nu_cgs, nu_cgs*Lnu_cgs, marker='o', ls='', label='Data');
plt.loglog(dense_freqs, fit_vals, ls='-', color='black', label=fit_string)
#plt.loglog(dense_freqs, nuLnu(dense_freqs, temp_fit, 5e-4), ls='-', color='black', label=fit_string)
plt.xlabel(r'$\nu$ [Hz]')
plt.ylabel(r'$\nu L_\nu$ [erg/s]')
plt.legend();
```

### d) Optical depth consistency check

Calculate $\alpha_\nu R$ for your favored parameters assuming LTE. Is the result consistent with our assumption in part c? Assuming $R$ is a constant, at what frequency $\nu_\tau$ should the bremsstrahlung radiation become optically thick (ignoring all other sources of emission, absorption, scattering)?

```python
# I ended up doing this mostly analytically, in the RJ limit. See pdf.
alpha_nu_const = 1e-38/(2*h_cgs/c_cgs**2)
alpha_nu = alpha_nu_const*density_fit**2/np.sqrt(temp_fit)
nu_cutoff = kb_cgs*temp_fit/h_cgs
print("Cutoff frequency: {:.3e} Hz".format(nu_cutoff))
tau_cutoff = radius_cgs*alpha_nu*(1.0 - np.exp(-h_cgs*nu_cutoff/(kb_cgs*temp_fit)))/nu_cutoff**3
print("Optical depth at cuoff frequency: {:.3e}".format(tau_cutoff))
nu_low = 1e1
nu_high = nu_cgs[-1]
tau_low = radius_cgs*alpha_nu*(1.0 - np.exp(-h_cgs*nu_low/(kb_cgs*temp_fit)))/nu_low**3
tau_high = radius_cgs*alpha_nu*(1.0 - np.exp(-h_cgs*nu_high/(kb_cgs*temp_fit)))/nu_high**3
print("Optical depth for nu low: {:.3e}".format(tau_low))
print("Optical depth for nu high: {:.3e}".format(tau_high))

print("{:.2e}".format(np.sqrt(6e21*h_cgs/kb_cgs/temp_fit)))

def tauv(nu):
    optically_thick_tau = 100
    tau = radius_cgs*alpha_nu*(1.0 - np.exp(-h_cgs*nu/(kb_cgs*temp_fit)))/nu**3
    return tau - optically_thick_tau

from scipy.optimize import fsolve
from scipy.optimize import root
# root(tauv, x0=[1e10])
tau_solution = fsolve(tauv, 1e20)
print(tau_solution)

print("{:.2e}".format(1.e-38*radius_cgs*density_fit**2/(2.*kb_cgs*temp_fit**1.5/c_cgs**2)))
```

### c) Eddington limit

A minimum opacity value comes from electron (Thomson) scattering off free electrons, $\kappa = \kappa_T$, where $\kappa_T = \sigma_T / m_p = 0.4$ cm$^{2}$ g$^{-1}$ for pure hydrogen. Using this opacity, calculate the Eddington luminosity $L_{\rm Edd} (M)$, above which hydrogen will be ejected by radiation pressure. Scale your result to the mass of the Sun, $M_{\rm Sun} = 2\times10^{33}$ g.

```python
kappa_T = 0.4 # cm^2/g
Ledd = 4.0*np.pi*G_cgs*Msun_cgs*c_cgs/kappa_T
print("Eddington Luminosity for mass M in solar masses: {:.2e} M erg/s".format(Ledd))
```
